# Entropy, the beta distribution, and skew - Homework

As you know, the beta distribution Beta(a,b) is a continuous distribution on [0,1] that can be either symmetric or skew depending on the values of a and b. Let's take a look at the entropy of this distribution as a funciton of a and b. There is a formula for this, but it is somewhat complicated and it's easy enough to calculate this approximately using Julia's Distributions.jl. Let a and b each range from [1,5] with 9 equally spaced points for each (i.e. 1, 1.5, 2, 2.5, etc.). For each pair (a,b), create the distribution $\beta(a,b)$ and compute its entropy (I'll leave it to you to search how using Distributions.jl, it's very easy).

You now have an entropy value for Beta(a,b) for 81 pairs of (a,b). Plot a heatmap of these entropies. What can you conclude about how the shape of the beta distribution influences the entropy?

```julia
using Pkg
Pkg.add("Distributions")
using Distributions
using Plots
```

```julia
a = Vector(range(1, stop=5, length=9))
b = Vector(range(1, stop=5, length=9))
```

```julia
function beta_entropy_sim(a, b)
    beta_dists = []
    for (a,b) in zip(a,b)
        push!(beta_dists, Beta(a, b))
    end
    entropies = [entropy(dist) for dist in beta_dists]
   
    return entropies
end
```

```julia
entropy_mat = zeros(Float64, length(a), length(b))
beta_entropy_sim(a, b)
#Fill in entropy_mat with the entropies of the beta distributions
for i in 1:length(a)
    for j in 1:length(b)
        entropy_mat[i,j] = beta_entropy_sim([a[i]], [b[j]])[1]
    end
end
```

```julia
heatmap(a, b, entropy_mat, xlabel="a", ylabel="b", color=:coolwarm, legend=false,colorbar=:right)
```

```julia
#We can see that the entropy of the beta distribution increases when a and b is increased. 
#This is because the beta distribution becomes more spread out as a and b increase, and the entropy of a distribution is a measure of its spread/uncertainty.
```
