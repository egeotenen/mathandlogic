
"""
In this problem, you will work with the beta-binomial model, which is a common way
    of modeling probabilistic beliefs.

Suppose you have a political candidate and you are running their polling. Past polls
    have indicated 55-65% support. On the basis of these past polls, you assign the
    distribution Beta(40,27) as the probabilistic belief on support levels.

A new poll just came in that found 108/200 (54%) of participants support your candidate.
    We'll assume that this can be well modeled by a standard binomial distribution.

On the basis of this new poll, how should you update your distribution of possible support
    levels?

Part 1:
Defne the appropriate prior distribution using Distributions.jl. Plot this distribution
    and output the mean and standard eviation of that prior distribution. 
    This is your prior expectation and uncertainty.

Part 2:
Calculate the posterior distribution for this problem. Overlay a plot of the posterior
    on the same axes as the prior. Now calculate the mean and standard deviation of the
    posterior.

How much did this new poll change your expected level of support for the candidate?

Note that this is over-simplified and poll modeling is much, much more complex.
    These are the basic building blocks though!

"""

```julia
using Pkg
Pkg.instantiate()
Pkg.status()
```

```julia
using Distributions
using StatsPlots
```

```julia
using Distributions
using Plots

# Beta parameters
alpha_prior = 40
beta_prior = 27

# Mean and standard deviation for the prior distribution
mean_prior = alpha_prior / (alpha_prior + beta_prior)
std_dev_prior = sqrt(alpha_prior * beta_prior / ((alpha_prior + beta_prior)^2 * (alpha_prior + beta_prior + 1)))

# Make prior distribution
prior_distribution = Beta(alpha_prior, beta_prior)

# Generate x values for plotting
x = 0:0.001:1

# Plot the prior distribution
plot(x, pdf.(prior_distribution, x), label="Prior Distribution (Beta(40,27))", color=:blue)

# Additions for posterior distribution calculation based on the new poll
alpha_posterior = alpha_prior + 108  # successes 
beta_posterior = beta_prior + 92  # failures 




# Mean and standard deviation for the posterior distribution
mean_posterior = alpha_posterior / (alpha_posterior + beta_posterior)
std_dev_posterior = sqrt(alpha_posterior * beta_posterior / ((alpha_posterior + beta_posterior)^2 * (alpha_posterior + beta_posterior + 1)))

# Create the posterior distribution
posterior_distribution = Beta(alpha_posterior, beta_posterior)

# Plot the posterior distribution on the same axes as the prior
plot!(x, pdf.(posterior_distribution, x), label="Posterior Distribution (after new poll)", color=:red)

# Display the plot
display(plot)

# Print the mean and standard deviation for both distributions
println("Prior Distribution: Mean = $mean_prior, Standard Deviation = $std_dev_prior")
println("Posterior Distribution: Mean = $mean_posterior, Standard Deviation = $std_dev_posterior")


```

```julia
# If you define a prior and posterior distribution correctly using Distributions.jl
# these lines will plot them for you.


plot(prior_distribution,xlims=(0,1),color = "black",label = "prior")
plot!(posterior_distribution,xlims=(0,1),color = "red",label = "posterior")

#You can see distributions below
```
