# Binomial entropy calculation

In this problem, you will explore the entropy of the binomial distribution as a function of generating probability p.

To be precise, consider a single coin flip with p(heads) =  q. Create a function that computes the entroppy of this coin flip process (using a base 'e' log). Now, using this function, plot the entropy of this process as a function of q.

In calculating the entropy, you should code your own entropy calculation for the binomial. You cannot use a built in entropy() function. It's easy and only takes one line.

One note of caution. This calculation can become problematic at q=0.0, 1.0. I will not tell you why, but it is easy to see why. So just make sure to be aware of it.

```julia
using Pkg
Pkg.add("Distributions")
using Distributions
using Plots
```

```julia
#p(heads) = q 
#p(tails) = 1 - q
```

```julia
#entropy of a binomial distribution

function binomial_entropy(q)
    if q == 0 || q == 1 #because log(0) is undefined, log(1) = 0
        return 0
    end
    return -q*log(q) - (1-q)*log(1-q)
end
```

```julia
#defining a range of q values
q_values = Vector(range(0.01, stop=0.99, length=50))
```

```julia
#try if it brings meaniingful results (e.g., high entropy for q = 0.5)
binomial_entropy(0.5)
```

```julia
#check if the entropy is symmetric
for q in q_values
    println(binomial_entropy(q))
end
```

```julia
# My plot
plot(q_values, binomial_entropy.(q_values))
xlabel!("p")
ylabel!("Entropy")
```
