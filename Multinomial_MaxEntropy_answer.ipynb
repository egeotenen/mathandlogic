1) Create a multinomial distribution assuming the outcomes are equally likely. Then compute its entropy. This is a concrete number that will be used for comparison.

    Choose a collection of 6 random probabilities representing the 6 outcomes (note these 6 probabilities must add to 1; they are probabilities after all). For this random (but non uniform) distribution, compute the entropy.

```julia
# Defining the entropy function
function entropy(p)
    return -sum(p .* log.(p))
end

# Entropy of the non-uniform distributions
num_replications = 100_000
entropies = zeros(num_replications)

for i in 1:num_replications
    dirichlet_dist = Dirichlet(ones(6)) #This is the uniform distribution
    random_probs = rand(dirichlet_dist)    
    # Computing the entropy of the random set of probs
    entropies[i] = entropy(random_probs)
end

# Plotting the histogram of the entropies
histogram(entropies, bins=100, alpha=0.6, label="Random Probabilities Entropy", xlabel="Entropy", ylabel="Frequency", title="Entropy Distribution for Random Probability Sets")
vline!([comparison_ent], color=:red, linewidth=2, label="Uniform Entropy")

```
